```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'C:/Users/alish/CIA Mayur Sir/train.csv'
train_data = pd.read_csv(file_path)
```

```python
numeric_desc = train_data.describe()
print("Numeric Variables Descriptive Statistics:")
display(numeric_desc)
```

```python
categorical_desc = train_data.describe(include=['object'])
print("\nCategorical Variables Descriptive Statistics:")
display(categorical_desc)
```

# Graphical Analysis

# Age Distribution
This plot shows the distribution of ages among the passengers. The histogram will have bins representing age ranges, and the height of each bin indicates the number of passengers within that range. The KDE (Kernel Density Estimate) line will smooth out the distribution, giving you an idea of the density of ages.

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))
sns.histplot(train_data['Age'].dropna(), bins=30, kde=True, color='steelblue', line_kws={'color': 'darkblue', 'lw': 2})
plt.title('Age Distribution of Passengers', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

# Class Distribution and Survival Rates
Interpretation:
Class Distribution: Most passengers are in 3rd class, followed by 1st and 2nd classes.
Survival Rates by Class: The survival rate is significantly higher for 1st class passengers compared to 3rd class. This reflects the higher priority given to passengers in 1st class during evacuation.


```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=train_data, palette='Set2')  
plt.title('Number of Passengers by Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=train_data, palette='Set2', ci=None)  
plt.title('Survival Rate by Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

# Gender Distribution and Survival Rates
Interpretation:
Gender Distribution: There are more male passengers than female passengers.
Survival Rates by Gender: The survival rate for females is significantly higher than for males, which reflects the "women and children first" policy during the evacuation.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=train_data, palette='Paired')  
plt.title('Number of Passengers by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=train_data, palette='Paired', ci=None)  # Remove the confidence interval
plt.title('Survival Rate by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()

```

# Number of Passengers by Class
This bar plot displays the count of passengers in each class (Pclass). The x-axis will have the class labels (1, 2, 3), and the y-axis will show the number of passengers in each class. It helps visualize the distribution of passengers across the different classes.

#  Age by Class
This box plot shows the distribution of ages for each passenger class. Each box represents the interquartile range (IQR) of ages for that class, with the line inside the box indicating the median age. Whiskers extend to show the range, and any points outside the whiskers are considered outliers. This plot helps in understanding how age varies across classes.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Age', data=train_data)
plt.title('Age Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

# Pair Plot for Numeric Variables
This pair plot shows scatter plots for every pair of numeric variables (Age, SibSp, Parch, Fare). The diagonal will have histograms of each numeric variable. The scatter plots off the diagonal help in visualizing relationships and correlations between the numeric variables.

```python
plt.figure(figsize=(12, 10))
sns.pairplot(train_data[['Age', 'SibSp', 'Parch', 'Fare']].dropna())
plt.show()
```
